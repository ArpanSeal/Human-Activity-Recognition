```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.svm import SVC
import joblib
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Load the CSV file into a pandas dataframe
df = pd.read_csv('HAR_new.csv')

# Remove rows containing NaN values
df.dropna(inplace=True)

# Add 'Activity' column based on 'aMagnitude' values
df['Activity'] = pd.cut(df['aMagnitude'], bins=[0, 1, 4.5, float('inf')], labels=['NO_ACTIVITY', 'WALKING', 'JOGGING'])

# Sort the dataframe by 'Activity' column
df = df.sort_values(by='Activity')

# Save the modified dataframe to a new CSV file
df.to_csv('modified_data.csv', index=False)
```

```python
# Read the csv file
df = pd.read_csv('modified_data.csv')

new_df = df[['ax', 'ay', 'az', 'aMagnitude', 'Activity']]

new_df.insert(0, 'time', range(0, len(new_df)))

new_df.to_csv('acceleration.csv', index=False)
```

```python
# Extract the 'ax' and 'time' columns from the dataframe
AX = new_df['ax']
time = new_df['time']

# Create a plot
fig, ax = plt.subplots(figsize=(11, 7))
ax.plot(time, AX)
ax.set_xlabel('Time')
ax.set_ylabel('Aceeleration-X')

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='r', labelsize='medium', width=5)

# Show the plot
plt.show()
```

```python
# Read the csv file
df = pd.read_csv('acceleration.csv')

# Select the rows where Activity is NO_ACTIVITY, WALKING, and JOGGING
no_activity = df[df['Activity'] == 'NO_ACTIVITY']
walking = df[df['Activity'] == 'WALKING']
jogging = df[df['Activity'] == 'JOGGING']

# Create subplots with 3 rows and 1 column
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))

# Plot ax vs time for each activity
ax[0].plot(no_activity['time'], no_activity['ax'], color='#E64A19')
ax[1].plot(walking['time'], walking['ax'], color='#1E88E5')
ax[2].plot(jogging['time'], jogging['ax'], color='#689F38')

# Set the x and y labels
ax[0].set_xlabel('Time', fontsize=14)
ax[0].set_ylabel('Acceleration-X (m/s^2)', fontsize=14)
ax[1].set_xlabel('Time', fontsize=14)
ax[1].set_ylabel('Acceleration-X (m/s^2)', fontsize=14)
ax[2].set_xlabel('Time', fontsize=14)
ax[2].set_ylabel('Acceleration-X (m/s^2)', fontsize=14)

# Set the titles for each subplot
ax[0].set_title('NO_ACTIVITY', fontsize=16, fontweight='bold')
ax[1].set_title('WALKING', fontsize=16, fontweight='bold')
ax[2].set_title('JOGGING', fontsize=16, fontweight='bold')

# Add gridlines to the plots
ax[0].grid(True)
ax[1].grid(True)
ax[2].grid(True)

# Set the tick font size
ax[0].tick_params(labelsize=12)
ax[1].tick_params(labelsize=12)
ax[2].tick_params(labelsize=12)

# Adjust the subplot spacing
plt.subplots_adjust(hspace=0.5)

# Show the plots
plt.show()
```

```python
train = pd.read_csv('modified_data.csv')

# Define the feature columns and target column
feature_column = train.columns[:-1]
target_column = 'Activity'

x_train = train[feature_column]
y_train = train[target_column]

# Train the linear SVM classifier
linear_svm = SVC(kernel='rbf', C=1.0, gamma='auto')
linear_svm = linear_svm.fit(x_train, y_train)

# Save the trained classifier as a pickle file
joblib.dump(linear_svm, 'svm_model_new.pkl')

y_pred = linear_svm.predict(x_train)
```

```python
print('Accuracy:', accuracy_score(y_train, y_pred))

print()

# Calculate accuracy
accuracy = accuracy_score(y_train, y_pred)

# Generate classification report
classification_rep = classification_report(y_train, y_pred, output_dict=True)

# Create a DataFrame from classification report
df = pd.DataFrame(classification_rep).transpose()

# Display the output table
print(df)
```

```python
# Define the list of labels for the confusion matrix
labels = ['NO_ACTIVITY', 'WALKING', 'JOGGING']

# Compute the confusion matrix for the predicted labels and store it
confusion_mat = confusion_matrix(y_train, y_pred)

# Create a figure and axis for the plot
figure, axis = plt.subplots()
figure.set_size_inches(8, 6)

# Set the title for the plot
axis.set(title='Confusion Matrix for Linear SVM')

# Generate a heatmap of the confusion matrix
sns.heatmap(confusion_mat, ax=axis, annot=True, fmt='d', cmap='Purples', xticklabels=labels, yticklabels=labels)

# Automatically adjust the spacing between subplots to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
